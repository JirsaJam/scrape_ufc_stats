```python
import pandas as pd

# List of columns to process
landed_columns = [
    'SIG.STR._Landed', 'TOTAL STR._Landed', 'TD_Landed',
    'HEAD_Landed', 'BODY_Landed', 'LEG_Landed',
    'DISTANCE_Landed', 'CLINCH_Landed', 'GROUND_Landed'
]

# Corresponding attempted columns
attempted_columns = [
    'SIG.STR._Attempted', 'TOTAL STR._Attempted', 'TD_Attempted',
    'HEAD_Attempted', 'BODY_Attempted', 'LEG_Attempted',
    'DISTANCE_Attempted', 'CLINCH_Attempted', 'GROUND_Attempted'
]

def calculate_avoided(df, landed_col, attempted_col):
    for _, group in df.groupby(['EVENT', 'BOUT', 'ROUND']):
        if len(group) != 2:
            continue  # Skip if we don't have exactly 2 fighters in the group (sanity check)

        fighter_1 = group.iloc[0]
        fighter_2 = group.iloc[1]

        df.loc[group.index[0], f'{attempted_col}_avoided'] = fighter_2[attempted_col] - fighter_2[landed_col]
        df.loc[group.index[1], f'{attempted_col}_avoided'] = fighter_1[attempted_col] - fighter_1[landed_col]


# Apply the function to each pair of columns
for landed_col, attempted_col in zip(landed_columns, attempted_columns):
    calculate_avoided(df, landed_col, attempted_col)

# Fill any NaN values with 0 (for cases where there's no opponent in the next row)
df.fillna(0, inplace=True)

# The DataFrame df now has "avoided" columns for each attempted/landed pair, reflecting what the fighter avoided
```

```python
def calculate_received(df, landed_col):
    # Group by EVENT, BOUT, ROUND to get pairs of fighters
    for _, group in df.groupby(['EVENT', 'BOUT', 'ROUND']):
        if len(group) != 2:
            continue  # Skip if we don't have exactly 2 fighters in the group (sanity check)

        # Fighter 1 and Fighter 2
        fighter_1 = group.iloc[0]
        fighter_2 = group.iloc[1]

        # Calculate received strikes
        df.loc[group.index[0], f'{landed_col}_received'] = fighter_2[landed_col]
        df.loc[group.index[1], f'{landed_col}_received'] = fighter_1[landed_col]

# Apply the function to each landed column
for landed_col in landed_columns:
    calculate_received(df, landed_col)
```

```python
import pandas as pd

# Initialize Elo ratings for fighters
elo_ratings = {}

# Define constants
K = 32  # You can adjust this value based on how volatile you want the ratings to be

# Function to calculate expected score
def expected_score(rating, opponent_rating):
    return 1 / (1 + 10 ** ((opponent_rating - rating) / 400))

# Function to update Elo ratings
def update_elo(winner_rating, loser_rating):
    winner_expected = expected_score(winner_rating, loser_rating)
    loser_expected = expected_score(loser_rating, winner_rating)
    
    winner_new_rating = winner_rating + K * (1 - winner_expected)
    loser_new_rating = loser_rating + K * (0 - loser_expected)
    
    return winner_new_rating, loser_new_rating

# Reset the index to ensure uniqueness
elos = elos.reset_index(drop=True)

# Create columns to store the post-fight Elo ratings
elos['ELO_FIGHTER'] = 0.0

# Iterate over the dataset by identifying the winner first
for index, row in elos.iterrows():
    fighter = row['FIGHTER']
    outcome = row['RESULT']
    
    # Initialize ratings if not present
    if fighter not in elo_ratings:
        elo_ratings[fighter] = 1500
    
    # Identify opponent based on the unique Bout and Event
    opponent_row = elos[(elos['EVENT'] == row['EVENT']) & 
                        (elos['BOUT'] == row['BOUT']) & 
                        (elos['FIGHTER'] != fighter)]
    opponent = opponent_row['FIGHTER'].values[0]
    
    if opponent not in elo_ratings:
        elo_ratings[opponent] = 1500
    
    # Determine who won and lost
    if outcome == 'Win':
        winner, loser = fighter, opponent
    else:
        winner, loser = opponent, fighter
    
    # Update ratings after the fight
    winner_new, loser_new = update_elo(elo_ratings[winner], elo_ratings[loser])
    
    # Assign new ratings
    elo_ratings[winner] = winner_new
    elo_ratings[loser] = loser_new
    
    # Store the updated Elo ratings after the fight for both fighters
    elos.at[index, 'ELO_FIGHTER'] = elo_ratings[fighter]
    elos.at[opponent_row.index[0], 'ELO_FIGHTER'] = elo_ratings[opponent]

elo_fin = elo_ratings
```

    'Decision - Unanimous': 1.0,

    winner_expected = expected_score(winner_rating, loser_rating)
    loser_expected = expected_score(loser_rating, winner_rating)
    
    winner_new_rating = winner_rating + K * modifier * (1 - winner_expected)
    loser_new_rating = loser_rating + K * modifier * (0 - loser_expected)
    
    return winner_new_rating, loser_new_rating

```python
# Define the additional delta metrics
additional_delta_metrics = {
    'delta_SIG_STR_vs_avoid': ('CUM_SIG.STR._Landed_PM', 'CUM_SIG.STR._Attempted_avoided_PM'),
    'delta_TD_vs_avoid': ('CUM_TD_Landed_PM', 'CUM_TD_Attempted_avoided_PM'),
}

# Create a list to hold the new delta columns
delta_columns = []

# Iterate over each bout by grouping the DataFrame
for (event, bout), group in df_with_deltas.groupby(['EVENT', 'BOUT']):
    if len(group) == 2:  # Ensure we have exactly two fighters in each bout
        fighter1 = group.iloc[0]
        fighter2 = group.iloc[1]
        
        # Calculate additional deltas for both fighters
        for delta_name, (metric1, metric2) in additional_delta_metrics.items():
            # Fighter 1 relative to Fighter 2
            delta_fighter1 = fighter1[metric1] - fighter2[metric2]
            df_with_deltas.loc[fighter1.name, delta_name] = delta_fighter1
            
            # Fighter 2 relative to Fighter 1
            delta_fighter2 = fighter2[metric1] - fighter1[metric2]
            df_with_deltas.loc[fighter2.name, delta_name] = delta_fighter2
            
            # Track the delta column
            delta_columns.append(delta_name)
```
